```python
# data manipulation
import pandas as pd
import numpy as np
import scipy.io.arff

# data visualization
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# carregando o dataset
data, meta = scipy.io.arff.loadarff(f'data/speeddating.arff')
df = pd.DataFrame(data)

# Convertendo byte-strings para strings
str_df = df.select_dtypes([object]).stack().str.decode('utf-8').unstack()
for col in str_df:
    df[col] = str_df[col]

# ajeitar o tipo dos dados
nominal_cols = [col for col, dtype in zip(meta.names(), meta.types()) if dtype == 'nominal']
for col in nominal_cols:
    df[col] = df[col].astype('category')

df.head()
```

```python
df.info(verbose=True)
```

```python
missing_values = 0
col_missing_values = 0

for col in df.columns:
    valores_ausentes = df[col].isna().sum()
    if valores_ausentes > 0:
        col_missing_values += 1
        print(f"{col} tem {valores_ausentes} valores ausentes")
    missing_values += valores_ausentes
```

```python
numeric_columns = df.select_dtypes(include=['number'])
# dropando colunas nao muito interessantes ou que os dados se repetem, como pref_o_atributo e atributo_important
numeric_columns.drop(
    columns=[
        'wave', 'age_o', 'd_age', 'interests_correlate', 'attractive_important',
        'sincere_important', 'intellicence_important', 'funny_important',
        'ambtition_important', 'shared_interests_important'
        ],
      inplace=True
    )
```

```python
df_melted = numeric_columns.melt(var_name="Column", value_name="Value")

plt.figure(figsize=(12, 6))

# boxplot
sns.boxplot(x="Column", y="Value", data=df_melted, palette='coolwarm', flierprops={"marker": "o", "markerfacecolor": "red", "markersize": 6})

plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.title("Boxplot")
plt.show()
```

Aqui vemos que existem muitos possÃ­veis outliers (nÃ£o necessariamente sÃ£o outliers, Ã© necessÃ¡rio fazer uma anÃ¡lise com conhecimento de domÃ­nio). Acreditamos que esses valores nÃ£o sÃ£o inconsistÃªncias, com exceÃ§Ã£o de gaming e reading (que passam de 10, quando deveriam ser valores entre 1 e 10).

```python
threshold = 0.5

corr_matrix = numeric_columns.corr()

masked_corr = corr_matrix.copy()
np.fill_diagonal(masked_corr.values, np.nan)

significant_columns = masked_corr.columns[
    (masked_corr.abs() > threshold).any(axis=0)
]

filtered_corr_matrix = corr_matrix.loc[significant_columns, significant_columns]

plt.figure(figsize=(12, 10))
mask = np.triu(np.ones_like(filtered_corr_matrix, dtype=bool))
sns.heatmap(
    filtered_corr_matrix,
    mask=mask,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    linewidths=0.5,
    square=True,
    annot_kws={'size':10}
)

plt.title("Heatmap das CorrelaÃ§Ãµes dos Atributos", fontsize=16)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

Aqui podemos ver que temos algumas colunas relativamente bem correlacionadas, sendo talvez interessante aplicar alguma forma de tratamento para diminuir a quantidade de colunas muito correlacionadas, como Principal Component Analysis.
